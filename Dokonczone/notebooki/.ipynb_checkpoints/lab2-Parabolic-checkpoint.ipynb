### Równanie paraboliczne

$$ u_{t} = u_{xx} \,\,\,\,\,\, x \in (0, 1), \,\,\,\, t > 0 $$

$$ u(t, 0) = u(t, 1) = 0 \,\,\,\,\,\, t>0 $$

$$ u(0, x) = \sin(\pi x) \,\,\,\,\,\, x \in [0, 1] $$

Rozwiązanie dokładne: $$ u(t, x) = e^{-\pi^2 t} \sin(\pi x) $$

```python
import numpy as np
import scipy as sc
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
from typing import Optional

%matplotlib notebook
from jupyterthemes import jtplot
jtplot.style(theme='monokai', context='notebook', ticks=True, grid=False)
```

```python
class SetupParabolic:
    def __init__(self,
                 dx: float,
                 dt: Optional[float] = None,
                 x_range = (0,1),
                 t_range = (0,1),
                 boundary_condition_low = lambda t: 0,
                 boundary_condition_high = lambda t: 0,
                 initial_condition = lambda x: 0,
                 output_freq: int = 100,
                 alpha: float = 0
                 ):
        
        self.x_range = x_range
        self.t_range = t_range
        self.boundary_condition_low = boundary_condition_low
        self.boundary_condition_high = boundary_condition_high
        self.initial_condition = initial_condition
        self.alpha = alpha  # just ignore it
        dt = dt or 0.5*(dx ** 2)

        # There are 2 hard problems in computer science:
        # cache invalidation, naming things, and off-by-1 errors.
        self.x_num = round((self.x_range[1] - self.x_range[0]) / dx) + 1
        self.t_num = round((self.t_range[1] - self.t_range[0]) / dt) + 1

        self.X, self.dx = np.linspace(*self.x_range, self.x_num, retstep=True)
        self.T, self.dt = np.linspace(*self.t_range, self.t_num, retstep=True)

        self.output_freq = output_freq
        
```

```python
from scipy.sparse import dia_matrix, identity
def Lh(setup):
    n = setup.x_num-2
    diag = np.ones(n)
    return (1/setup.dx**2)*dia_matrix(([diag, -2*diag, diag], [-1,0,1]), shape=(n,n))    
```

```python
def scheme_parabolic(setup):
    s = setup.dt/(setup.dx**2)    
    u_matrix = np.zeros((setup.x_num, setup.t_num // setup.output_freq))
    u_matrix[:,0] = setup.initial_condition(setup.X)
    u_old = u_matrix[:,0]
    u_new = np.zeros(setup.x_num)
    
    if setup.alpha == 0:
        # jesli alpha = 0, to nie trzeba odwracac macierzy
        for t in range(2, setup.t_num):  # suboptimal, but simple

            u_new[0] = setup.boundary_condition_low(setup.T[t])
            u_new[1] = (1-2*s)*u_old[1]+s*u_old[2] + s*setup.boundary_condition_low(setup.T[t])
            for i in range(2,setup.x_num-2):
                u_new[i] = s*(u_old[i-1] + u_old[i+1]) + (1-2*s)*u_old[i]
            u_new[setup.x_num-2] = s*u_old[setup.x_num-3] + (1-2*s)*u_old[setup.x_num-2] + s*setup.boundary_condition_high(setup.T[t])
            u_new[setup.x_num-1] = setup.boundary_condition_high(setup.T[t])
            if (t-1) % setup.output_freq == 0:
                u_matrix[:, (t-1) // setup.output_freq] = u_new[:]
            u_old = u_new
            
    else:
       # L = Lh(setup)
       # X = (1/setup.dt)*identity(setup.x_num-2) - setup.alpha*L
        M = sc.sparse.linalg.inv(((1/setup.dt)*identity(setup.x_num-2) - setup.alpha*Lh(setup)).tocsc())
        for t in range(2, setup.t_num):
            u_new[0] = setup.boundary_condition_low(setup.T[t])
            u_new[1:-1] = (1-1/setup.alpha)*u_old[1:-1] +(1/(setup.alpha*setup.dt))*M*u_old[1:-1]
            u_new[1] += (1/setup.dx**2)*(setup.alpha*setup.boundary_condition_low(setup.T[t])+(1-setup.alpha)*setup.boundary_condition_low(setup.T[t-1]))
            u_new[setup.x_num-2] += (1/setup.dx**2)*(setup.alpha*setup.boundary_condition_high(setup.T[t])+(1-setup.alpha)*setup.boundary_condition_high(setup.T[t-1]))
            u_new[setup.x_num-1] = setup.boundary_condition_high(setup.T[t])
            if (t-1) % setup.output_freq == 0:
                u_matrix[:, (t-1) // setup.output_freq] = u_new[:]
            u_old = u_new          
            
    
    return u_matrix
```

```python
def plot_surface(u, setup):
    fig = plt.figure()
    ax = plt.axes(projection='3d')
    # Prepare grid.
    T = np.linspace(*setup.t_range, setup.t_num // setup.output_freq)
    X = np.linspace(*setup.x_range, setup.x_num)
    T, X = np.meshgrid(T, X)

    # Plot the surface.
    surf = ax.plot_surface(T, X, u, cmap=cm.coolwarm,
                           linewidth=0, antialiased=False)

    # Customize the z axis.
    ax.zaxis.set_major_locator(LinearLocator(10))
    ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
    # Add a color bar which maps values to colors.
    fig.colorbar(surf, shrink=0.5, aspect=5)
    plt.show()
    #pass
    # TODO
```

```python
%%time
# numerical solution
setup_parabolic = SetupParabolic(
    dx=0.01,
    x_range = (0,1),
    t_range = (0,2),
    initial_condition = lambda x: np.sin(np.pi*x), # u(x,0) = sin(pi*x)
    boundary_condition_low = lambda t: 0,          # u(0,t) = 0
    boundary_condition_high = lambda t: 0,         # u(1,t) = 0
    output_freq = 100,
    alpha = 0.5
)
numerical_parabolic = scheme_parabolic(setup_parabolic)
```

```python
# exact solution
exact = lambda t,x: np.exp(-(np.pi**2)*t) * np.sin(np.pi*x)
X = np.linspace(*setup_parabolic.x_range, setup_parabolic.x_num)
T = np.linspace(*setup_parabolic.t_range, setup_parabolic.t_num // setup_parabolic.output_freq)
X, T = np.meshgrid(X, T)

exact_parabolic = exact(T, X).T
```

```python
plot_surface(numerical_parabolic, setup_parabolic)
plot_surface(exact_parabolic, setup_parabolic)
plot_surface(numerical_parabolic - exact_parabolic, setup_parabolic)
```

#### Błąd vs $\alpha$

```python
# exact solution
setup_parabolic = SetupParabolic(
    dx=0.05,
    x_range = (0,1),
    t_range = (0,2),
    initial_condition = lambda x: np.sin(np.pi*x), # u(x,0) = sin(pi*x)
    boundary_condition_low = lambda t: 0,          # u(0,t) = 0
    boundary_condition_high = lambda t: 0,         # u(1,t) = 0
    output_freq = 100,
    alpha = 0
)

exact = lambda t,x: np.exp(-(np.pi**2)*t) * np.sin(np.pi*x)
X = np.linspace(*setup_parabolic.x_range, setup_parabolic.x_num)
T = np.linspace(*setup_parabolic.t_range, setup_parabolic.t_num // setup_parabolic.output_freq)
X, T = np.meshgrid(X, T)
exact_parabolic = exact(T, X).T

alpha_axis = np.arange(0,1,0.05)
err_axis = []
for alpha in alpha_axis:
    setup_parabolic.alpha = alpha
    numerical_parabolic = scheme_parabolic(setup_parabolic)
    abs_error = np.abs(numerical_parabolic - exact_parabolic).max()
    err_axis.append(abs_error)

fig, ax = plt.subplots(figsize=(6,3), tight_layout=True)
ax.plot(alpha_axis, err_axis)
plt.show() 
```
